```python
df = pd.read_csv("world-happiness-report.csv")
df.head()
```

```python
#Filter for 2020 and sort to check out the top 10 happiest and least happy countries
top10 = df[df["year"] == 2020].sort_values("Life Ladder", ascending = False).head(n=10)
top10
```

```python
#Filter for 2020 and sort to check out the top 10 least happy countries
bot10 = df[df["year"] == 2020].sort_values("Life Ladder").head(n=10)
bot10
```

```python
top3 = ["Finland", "Iceland", "Denmark"]
bot3 = ["Zimbabwe", "Tanzania", "Jordan"]

#top 3 dataframe
df2 = df[df["Country name"].isin(top3)]
df2 = df2[df2["year"] > 2007]
df2
```

```python
#bottom 3 dataframe
df3 = df[df["Country name"].isin(bot3)]
df3 = df3[df3["year"] > 2007]
df3
```

```python
#Map stuff
dfmap = df.copy()

#ADAPTING TO THE ISO 3166 STANDARD
dfmap.loc[dfmap['Country name'] == 'Taiwan Province of China', 'Country name'] = 'Taiwan, Province of China' 
dfmap.loc[dfmap['Country name'] == 'Hong Kong S.A.R. of China', 'Country name'] = 'Hong Kong' 
dfmap.loc[dfmap['Country name'] == 'Congo (Brazzaville)','Country name'] = 'Congo' 
dfmap.loc[dfmap['Country name'] == 'Congo (Kinshasa)', 'Country name'] = 'Democratic Republic of the Congo'
dfmap.loc[dfmap['Country name'] == 'Palestinian Territories','Country name'] = 'Palestine, State of' 
dfmap.loc[dfmap['Country name'] == 'Somaliland region','Country name'] = 'Somalia' 

dfmap.drop(index=dfmap[dfmap['Country name'] == 'Kosovo'].index, inplace=True) 
dfmap.drop(index=dfmap[dfmap['Country name'] == 'North Cyprus'].index, inplace=True) 

#Add column to dfmap dataframe
dfmap['iso_alpha'] = dfmap['Country name'].apply(lambda x:pc.country_name_to_country_alpha3(x,))
```

```python
top10['iso_alpha'] = top10['Country name'].apply(lambda x:pc.country_name_to_country_alpha3(x,))
bot10['iso_alpha'] = bot10['Country name'].apply(lambda x:pc.country_name_to_country_alpha3(x,))
```

```python
#Create dataframe that combines top10 and bottom 10
top10["category"] = "top"
bot10["category"] = "bot"
topbot = top10.append(bot10)
topbot.tail()
```
